```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import csv
import re
from selenium.webdriver.common.action_chains import ActionChains

driver = webdriver.Chrome()

driver.get("https://www.seethroughny.net/payrolls")
i=0
while i<43:
    driver.maximize_window()
    element = driver.find_element_by_xpath('//div [@id="data_loader"]')
    driver.execute_script("return arguments[0].scrollIntoView(0, document.documentElement.scrollHeight-10);", element)
    actions = ActionChains(driver)
    actions.move_to_element(element)
    actions.click()
    actions.send_keys(webdriver.common.keys.Keys.SPACE)
    actions.perform()
    driver.implicitly_wait(15)
    SCROLL_PAUSE_TIME = 1
    time.sleep(SCROLL_PAUSE_TIME)
    i=i+1
    print('y')

csv_file = open('salary2.csv', 'w', encoding='utf-8')
writer = csv.writer(csv_file)
ele2=driver.find_element_by_xpath('//div [@id="results-search"]')
driver.execute_script("return arguments[0].scrollIntoView(0, document.documentElement.scrollHeight-10);", ele2)
reviews = driver.find_elements_by_xpath('//table[@class="table table-striped table-hover dt-responsive filter_results"]/tbody/tr[contains(@id, "resultRow")]')



print(len(reviews))
while True:
    try:
        for review in reviews:
            review_dict = {}
            expand_btn = review.find_element_by_xpath('./td[1]')
            expand_btn.click()
            name = review.find_element_by_xpath('./td[2]').text
            agency = review.find_element_by_xpath('./td[3]').text
            total_pay = review.find_element_by_xpath('./td[4]').text
#             print(name)
#             print(agency)
#             print(total_pay)

            attributes = review.find_elements_by_xpath('./following-sibling::tr[1]/td')
            attributes = [x.text for x in attributes[1:]]
            title=attributes[0].split('\n')[1]

            rateofpay=attributes[0].split('\n')[3]
            year=attributes[0].split('\n')[5]
            basis=attributes[0].split('\n')[7]
            branch=attributes[0].split('\n')[9]
#             print(rateofpay)
#             print('=' * 50)
            #time.sleep(1)
            review_dict['name']=name
            review_dict['agency']=agency
            review_dict['total_pay']=total_pay
            review_dict['title']=title
            review_dict['rate_of_pay']=rateofpay
            review_dict['year']=year
            review_dict['basis']=basis
            review_dict['branch']=branch
            writer.writerow(review_dict.values())
            
    except Exception as e:
        print(e)
        csv_file.close()
        driver.close()
        break
```
